## 1. group by region and save into separate .csv files 

We know that there are no missing values in the `Region` column, but do all regions represent significant groups? 
Let's check **total number of countries per region** (and **mean** `Happiness Score` while we're at it).

```python
df_grouped = df_cleaned.groupby('Region').agg({'Country': 'count', 'Happiness Score': np.mean})
df_grouped 
```

This shows that while all the other European countries either belong to the `Eastern Europe` or the `Western Europe` `region`, one unique country has been attributed to the `Europe region`.
Which one is that, and can we reallocate it to one of the other two regions?


```python
df_cleaned[df_cleaned.Region=='Europe']
```

Based on the geographical location of Ukraine, and Russia etc having been classified as belonging to `Eastern Europe`, we could reset the `Region` value of Ukraine to `Eastern Europe`.
Let's first create a copy of `df_cleaned` and then reset region value of Ukraine.

```python
df_cleaned_copy = df_cleaned.copy()
df_cleaned_copy.loc[73,'Region'] = 'Eastern Europe'
regions = df_cleaned_copy.Region.unique() #to check that we now have 6 and not 7 regions, and have a list of regions to iterate through below
regions
```

Let's use the `regions` variable that contains 6 unique values from `Region` column to filter the data frame for every region and save a separate `.csv` file with all the data for each region.

```python
def csv_by_group(df,grouper,column):
    for group in grouper:
        df_group = df[df[column]==group]
        df_group.to_csv(os.path.join('.', 'Datafolder', group+'.csv'))

csv_by_group(df_cleaned_copy,regions,'Region')
```

# G. Probabilities

The probability of a country being in Western Europe, given its Happiness score is greater than 6:
$P(Western Europe|score>6$) should be equal to:
$\frac{P(WesternEuropescore>6)}{P(score>6)}$

Let's first compute the overall probability of a randomly selected country having a Happiness score > 6:

```python
nb_countries_above6 = df_cleaned_copy[df_cleaned_copy['Happiness Score'] > 6].shape[0] #tot. number of countries with happiness score > 6
Pabove6 = nb_countries_above6 / df_cleaned_copy.shape[0] #probability given after division by tot. number of countries
Pabove6 #probability that a randomly selected country has a Happiness score > 6 : 44/151
```

Next we can compute the probability of a randomly selected country being both in Western Europe and having a score >6: 

```python
nb_countries_WE_above6 = df_cleaned_copy[(df_cleaned_copy['Happiness Score'] > 6) & (df_cleaned_copy.Region == 'Western Europe')].shape[0]
P_WEabove6 = nb_countries_WE_above6/df_cleaned_copy.shape[0]
P_WEabove6
```

And finally, the probability of $P(Western Europe|score>6$):

```python
print(P_WEabove6/Pabove6)
```
